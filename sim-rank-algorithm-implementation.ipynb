## Objective
- Implement the SimRank algorithm.
- Create a toy citation network.
- Compute similarity between every pair of nodes.



```

import networkx as nx
import numpy as np

# Create toy citation network (Directed Graph)
G = nx.DiGraph()

# Add citation edges (A cites B means A -> B)
edges = [
    ('A', 'C'),
    ('B', 'C'),
    ('A', 'D'),
    ('C', 'D'),
    ('D', 'E'),
    ('B', 'E')
]

G.add_edges_from(edges)

print("Nodes:", G.nodes())
print("Edges:", G.edges())
```

### Step 1: Implement the SimRank algorithm (as per the basic SimRank equation, section 4.2 of the attached paper).

```

def simrank(G, C=0.8, max_iter=10, tol=1e-4):
    nodes = list(G.nodes())
    n = len(nodes)
    sim = np.identity(n)

    for iteration in range(max_iter):
        prev_sim = sim.copy()

        for i in range(n):
            for j in range(n):
                if i == j:
                    continue

                in_i = list(G.predecessors(nodes[i]))
                in_j = list(G.predecessors(nodes[j]))

                if len(in_i) == 0 or len(in_j) == 0:
                    sim[i][j] = 0
                else:
                    s = 0
                    for u in in_i:
                        for v in in_j:
                            s += prev_sim[nodes.index(u)][nodes.index(v)]

                    sim[i][j] = (C * s) / (len(in_i) * len(in_j))

        # Convergence check
        if np.linalg.norm(sim - prev_sim) < tol:
            break

    return nodes, sim

nodes, similarity_matrix = simrank(G)

print("SimRank Similarity Matrix:")
print(similarity_matrix)
```

### Step 2: Find out the similarity between each pair of nodes in the network.

```

print("\nPairwise Similarities:")
for i in range(len(nodes)):
    for j in range(len(nodes)):
        if i != j:
            print(f"Similarity({nodes[i]}, {nodes[j]}) = {similarity_matrix[i][j]:.4f}")
```

### Conclusion


In this assignment, the SimRank algorithm was successfully implemented on a toy citation network. The similarity between each pair of nodes was computed based on the principle that two nodes are similar if they are cited by similar nodes. The results show that nodes sharing common in-neighbors have higher similarity values, while unrelated nodes have similarity close to zero. This demonstrates how SimRank captures structural similarity in directed networks and can be effectively used for citation analysis, recommendation systems, and similarity detection in large networks.
