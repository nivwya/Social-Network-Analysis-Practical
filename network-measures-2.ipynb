## Task 1: Create the Karate Club Network

```python

import networkx as nx
import matplotlib.pyplot as plt

G = nx.karate_club_graph()
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
```

## Task 2: Calculate Closeness Centrality of Each Node

```python

closeness_centrality = nx.closeness_centrality(G)
for node, value in closeness_centrality.items():
    print(f"Node {node}: {value:.4f}")
```

## Task 3: Calculate Betweenness Centrality of Each Node

```python

betweenness_centrality = nx.betweenness_centrality(G)
for node, value in betweenness_centrality.items():
    print(f"Node {node}: {value:.4f}")
```

## Task 4: Display the Karate Club Network

```python

plt.figure(figsize=(8,6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=600, edge_color='gray')
plt.title("Karate Club Network")
plt.show()
```

## Task 5: Closeness Centrality of Karate Club Network

```python

closeness_karate = nx.closeness_centrality(G)
for node, value in closeness_karate.items():
    print(f"Node {node}: {value:.4f}")
```

## Task 6: Betweenness Centrality of Karate Club Network

```python

betweenness_karate = nx.betweenness_centrality(G)
for node, value in betweenness_karate.items():
    print(f"Node {node}: {value:.4f}")
```

## Task 7: Identify the Most Important Node

```python

most_important_closeness = max(closeness_karate, key=closeness_karate.get)
most_important_betweenness = max(betweenness_karate, key=betweenness_karate.get)

print("Most important node based on closeness centrality:", most_important_closeness)
print("Most important node based on betweenness centrality:", most_important_betweenness)
```
