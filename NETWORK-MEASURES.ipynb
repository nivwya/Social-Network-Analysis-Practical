## Task 1: Calculate the degree of each node

```python

import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

G = nx.karate_club_graph()
degree_dict = dict(G.degree())
degree_dict
```

## Task 2: Plot graph with node size scaled by degree

```python

node_sizes = [degree_dict[node]*100 for node in G.nodes()]

plt.figure(figsize=(8,6))
nx.draw(G, with_labels=True, node_size=node_sizes, node_color='lightblue')
plt.title("Karate Club Graph (Node size proportional to Degree)")
plt.show()
```

## Task 3: Degree Distribution

```python

degrees = list(degree_dict.values())

plt.figure()
plt.hist(degrees, bins=range(min(degrees), max(degrees)+2), align='left')
plt.xlabel("Degree")
plt.ylabel("Number of Nodes")
plt.title("Degree Distribution")
plt.show()
```

## Task 4: Cumulative Degree Distribution

```python

import numpy as np

degree_counts = np.bincount(degrees)
cumulative = np.cumsum(degree_counts[::-1])[::-1]

plt.figure()
plt.plot(range(len(cumulative)), cumulative)
plt.xlabel("Degree")
plt.ylabel("Cumulative Count")
plt.title("Cumulative Degree Distribution")
plt.show()
```

## Task 5: Local Clustering Coefficient (Example Node)

```python

node = 0

lcc_nx = nx.clustering(G, node)

neighbors = list(G.neighbors(node))
k = len(neighbors)

edges_between_neighbors = 0
for i in neighbors:
    for j in neighbors:
        if i < j and G.has_edge(i, j):
            edges_between_neighbors += 1

lcc_manual = (2 * edges_between_neighbors) / (k * (k - 1))

lcc_nx, lcc_manual
```

## Task 6: Diameter, Density, and Average Path Length

```python

diameter = nx.diameter(G)
density = nx.density(G)
avg_path_length = nx.average_shortest_path_length(G)

diameter, density, avg_path_length
```
